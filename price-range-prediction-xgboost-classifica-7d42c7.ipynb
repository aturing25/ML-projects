```python
#Importing the necessary packages  

import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from graphviz import *
```

```python
data = pd.read_csv('../input/train.csv')
test=pd.read_csv('../input/test.csv')
#Splitting into X and y as per the required for Scikit learn packages
X, y = data.iloc[:,:-1], data.iloc[:,-1]
```

```python
ID_test=test['id']
test.drop('id',axis=1,inplace=True)
```

```python
a1=np.array(X['sc_h'])
b1=np.array(X['sc_w'])
t1=np.array(test['sc_h'])
t2=np.array(test['sc_w'])
t12=np.multiply(t1,t2)
t3=np.array(test['px_height'])
t4=np.array(test['px_width'])
t34=np.multiply(t3,t4)

a2=np.array(X['px_height'])
b2=np.array(X['px_width'])
ab1=np.multiply(a1,b1)
ab2=np.multiply(a2,b2)
X['sc_area']=list(ab1)
X['px_area']=list(ab2)
test['sc_area']=list(t12)
test['px_area']=list(t34)
var1=['sc_h','sc_w','px_height','px_width']
for v in var1:
    X.drop(v,axis=1,inplace=True)
    test.drop(v,axis=1,inplace=True)
test.head()
```

```python
X.head()
```

```python
c1=np.array(X['m_dep'])
ca1=np.multiply(c1,ab1)
c2=np.array(X['mobile_wt'])
c12=np.divide(ca1,c2)
X['density']=list(c12)

d1=np.array(test['m_dep'])
da1=np.multiply(d1,t12)
d2=np.array(test['mobile_wt'])
d12=np.divide(da1,d2)
test['density']=list(d12)

X.drop('m_dep',axis=1,inplace=True)
X.drop('mobile_wt',axis=1,inplace=True)

test.drop('m_dep',axis=1,inplace=True)
test.drop('mobile_wt',axis=1,inplace=True)

test.head()
```

```python
num_feat=X.select_dtypes(include=[np.number])
num_feat.columns
```

```python
var3=['blue','three_g','four_g','wifi','touch_screen','dual_sim']
for v in var3:
    X[v]=X[v].astype('category')
    test[v]=test[v].astype('category')
    
```

```python
var4=['battery_power','clock_speed','int_memory','ram','pc','n_cores','talk_time','sc_area','px_area']
for v in var4:
    X[v]=(X[v]-X[v].mean())/X[v].std()
    test[v]=(test[v]-test[v].mean())/test[v].std()
test.head()
```

```python
#Reading the data


#Splitting the dataset into training and testing
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.25, random_state=123)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear',C=100).fit(X_train,y_train)
scoretrain = clf.score(X_train,y_train)
scoretest  = clf.score(X_test,y_test)
```

```python
predict = clf.predict(test)
predict[:10]
```
